```python
import gzip
import gensim 
import logging
import gensim.models.keyedvectors as Word2Vec

with open('CategoryEntertainment.txt', 'r') as file:
    education_words = file.read().replace('\n', ' ')
education_words
```

```python
model = Word2Vec.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
education_words = education_words.split()
education_words
```

```python

for word in education_words:
    try:
        print(word)
        print(model.most_similar(word, topn=10))
    except KeyError:
        print(word, "not in list")
```
